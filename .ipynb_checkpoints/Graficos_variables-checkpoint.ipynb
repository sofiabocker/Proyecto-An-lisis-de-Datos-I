```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('C:/Users/Ashley/Documents/Universidad/I semestre 2025/An√°lisis de Datos/Proyecto/bank+marketing/bank/bank-full.csv', sep = ';')
```

```python
cuantitativas = df.select_dtypes(include=['int64', 'float64'])
for col in cuantitativas.columns:
    plt.figure(figsize=(8,4))
    sns.histplot(df[col], kde = True, color = 'skyblue', edgecolor = 'black')
    plt.title(f'Distribuci√≥n de {col}', fontsize = 14)
    plt.xlabel(col)
    plt.ylabel('Frecuencia')
    plt.grid(True, linestyle='--',alpha = 0.5)
    plt.tight_layout()
    plt.savefig(f"distribucion_{col}.png")
    plt.show()
```

```python
for col in cuantitativas.columns:
    plt.figure(figsize=(8,4))
    sns.boxplot(data = df, x = 'y', y = col, palette = 'coolwarm')
    plt.title(f'{col} seg√∫n respuesta del cliente', fontsize = 14)
    plt.xlabel('Respuesta de (y)')
    plt.ylabel(col)
    plt.grid(True, linestyle='--',alpha = 0.4)
    plt.tight_layout()
    plt.show()
```

```python
for col in cuantitativas.columns:
    plt.figure(figsize=(8,4))
    sns.violinplot(data = df, x = 'y', y = col, palette = 'pastel')
    plt.title(f'Distribuci√≥n de {col} seg√∫n respuesta del cliente',  fontsize = 14)
    plt.xlabel('Respuesta')
    plt.ylabel(col)
    plt.grid(True, linestyle='--',alpha = 0.4)
    plt.tight_layout()
    plt.savefig(f"distribuviolin_{col}.png")
    plt.show()
```

```python

plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='age', y='balance', hue='y', alpha=0.5)
plt.title('Relaci√≥n entre edad y balance seg√∫n la respuesta')
plt.xlabel('Edad')
plt.ylabel('Balance')
plt.grid(True)
plt.tight_layout()
plt.savefig(f"edad_balance_respuesta.png")
plt.show()
```

```python

# Seleccionar solo las columnas num√©ricas
df_numeric = df.select_dtypes(include=['int64', 'float64'])

# Calcular la matriz de correlaci√≥n
correlation_matrix = df_numeric.corr()

# Crear el mapa de calor
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Mapa de calor de correlaciones entre variables num√©ricas")
plt.tight_layout()
plt.savefig(f"mapa_color.png")
plt.show()
```

```python

plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='education', hue='y', order=df['education'].value_counts().index)
plt.xticks(rotation=45)
plt.title('Relaci√≥n entre educaci√≥n y respuesta del cliente')
plt.xlabel('Nivel educativo')
plt.ylabel('Cantidad de clientes')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='job', hue='y', order=df['job'].value_counts().index)
plt.xticks(rotation=45)
plt.title('Ocupaci√≥n y respuesta del cliente')
plt.xlabel('Ocupaci√≥n')
plt.ylabel('Cantidad de clientes')
plt.tight_layout()
plt.show()
```

```python
from statsmodels.graphics.mosaicplot import mosaic

mosaic(df, ['education', 'y'])
plt.title('Relaci√≥n entre educaci√≥n y respuesta del cliente')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.stripplot(data=df, x='job', y='age', hue='y', dodge=True, alpha=0.6)
plt.xticks(rotation=45)
plt.title('Distribuci√≥n de edad por ocupaci√≥n y respuesta')
plt.tight_layout()
plt.show()
```

```python


# Filtrar variables categ√≥ricas
categoricas = df.select_dtypes(include='object').columns

# Crear un gr√°fico para cada una
for col in categoricas:
    plt.figure(figsize=(8, 4))
    sns.countplot(data=df, x=col, order=df[col].value_counts().index, palette='pastel')
    plt.title(f'Distribuci√≥n de la variable categ√≥rica: {col}')
    plt.xlabel(col.capitalize())
    plt.ylabel('Frecuencia')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.savefig(f"categorica_{col}.png")
    plt.show()
```

```python
sns.boxplot(x=df['balance'])
plt.title('Boxplot de balance')
plt.xlabel('Balance')
plt.show()
```
