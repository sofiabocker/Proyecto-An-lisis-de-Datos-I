# Cargar base de datos

```python
# Importar Librerías
import os
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Obtener la ruta del directorio actual
current_directory = os.path.dirname(_file) if "file_" in locals() else os.getcwd()

# Construir la ruta al archivo xlsx
file_path = os.path.join(current_directory, 'base.xlsx')

# Cargar el archivo xlsx en un DataFrame
df = pd.read_excel(file_path)
```

# Ajuste Modelo Árboles de Decisión

```python
# Primer paso: dividir las columnas en dos tipos de variables: dependientes e independientes
var_indep = ['job', 'marital', 'education', 'default','balance','housing','loan','contact','day','month','duration','campaign','pdays','previous','poutcome']
X = df[var_indep] # Variables Independientes
# Al tener variables categóricas, se convierten a numéricas
X = pd.get_dummies(X, drop_first=True)

y = df.y # Variable Objetivo
# La variable dependiente 'y' es categórica como 'yes' o 'no', por lo que se convierte a valores binarios
y = y.map({'no': 0, 'yes': 1})
```

```python
# Segundo paso: Dividir base de datos en conjunto de entrenamiento y de prueba

# Se dividen X y y en los sets de entrenamiento y de prueba
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.model_selection import RandomizedSearchCV
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
from sklearn import tree
from tabulate import tabulate
import matplotlib.pyplot as plt


# Se usa un radio de 75:25, es decir, 75% de los datos va a ser usados en el modelo de entrenamiento y 25% para el modelo de prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=16)
```

```python
# versión que optimiza la métrica de accuracy sin parámetros
modelo = tree.DecisionTreeClassifier()

modelo_optimo = RandomizedSearchCV(estimator = modelo,
                                        param_distributions = dict(),
                                        n_iter = 10,
                                        cv = 5,
                                        verbose=  2,
                                        random_state = 42,
                                        n_jobs = -1, 
                                        scoring = None)

modelo_optimo.fit(X_train, y_train)

modelo = modelo_optimo.best_estimator_

modelo.fit(X_train, y_train)

print(f"Mejores hiperparámetros: {modelo_optimo.best_params_}")
print(f"Mejor puntuación: {modelo_optimo.best_score_}")
```

```python
# Graficar curva ROC
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize

y_pred = modelo.predict_proba(X_test)[:,1]
fpr, tpr, _ = roc_curve(y_test, y_pred)
roc_auc = auc(fpr, tpr)

plt.figure()
plt.plot(fpr, tpr, label="Curva ROC (AUC = {:.2f})".format(roc_auc))
plt.plot([0, 1], [0, 1], color='black', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Tasa Predicciones No Precisas')
plt.ylabel('Tasa Predicciones Precisas')
plt.title('Cruva ROC - Modelo Sin Parámetros')
plt.legend(loc='lower right')
plt.grid()
plt.show()
```

0.71 indica que el modelo tiene un comportamiento aceptable, pero puedo mejorarse. Tiene un “codo” pronunciado al principio, lo cual es deseable ya que sugiere que el modelo es capaz de captar un número considerable de positivos con pocos falsos positivos al inicio y está por encima de la línea punteada, lo que confirma que el modelo tiene valor predictvo.



```python
# versión que optimiza la métrica de accuracy, pero con parámetros

from scipy.stats import randint, uniform
param_distributions = dict(
    max_depth = randint(1, 5),
    ccp_alpha = uniform()
)

modelo_cal = tree.DecisionTreeClassifier()
modelo_optimo_cal = RandomizedSearchCV(estimator = modelo_cal,
                                        param_distributions = param_distributions,
                                        n_iter = 10,
                                        cv = 5,
                                        verbose=  2,
                                        random_state = 42,
                                        n_jobs = -1, 
                                        scoring = None)

modelo_optimo_cal.fit(X_train, y_train)

modelo_cal = modelo_optimo_cal.best_estimator_

modelo_cal.fit(X_train, y_train)

print(f"Mejores hiperparámetros: {modelo_optimo_cal.best_params_}")
print(f"Mejor puntuación: {modelo_optimo_cal.best_score_}")
```

```python
# Graficar curva ROC
y_pred_prob = modelo_cal.predict_proba(X_test)[:,1]
fpr_cal, tpr_cal, _ = roc_curve(y_test, y_pred_prob)
roc_auc_cal = auc(fpr_cal, tpr_cal)

plt.figure()
plt.plot(fpr_cal, tpr_cal, label="Curva ROC (AUC = {:.2f})".format(roc_auc_cal))
plt.plot([0, 1], [0, 1], color='black', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Tasa Predicciones No Precisas')
plt.ylabel('Tasa Predicciones Precisas')
plt.title('Curva ROC - Con Parámetros')
plt.legend(loc='lower right')
plt.grid()
plt.show()
```

0.84 significa que tiene buena capacida discriminativa, comparado con el anterior distingue mucho mejor entre positivos y negativos. La curva se acerca mucho al borde superior, lo cual es ideal e indica que logra una alta tasa de verdaderos positivos, sin elevar demasiado la tasa de falsos positivos. 
El optimizador probablemente encontró hiperparámetros que generalizan mejor y no sobreajustan al conjunto de entrenamiento, lo que puede ser indicativo de que el dataset tiene desequilibrio de clases, y optimizar por accuracy no penalizaba adecuadamente los falsos positivos.
